Importar las librerias necesarias:

```python
import pandas as pd
import pyspc as ps
from pyspc import *
```

Leer el dataset "weatherAUS" desde el ordenador

```python
climaAUS_df= pd.read_csv("weatherAUS.csv")
climaAUS_df.head()
```

Separar solo los primeros 500 datos del DataSet, debido a que son más de 50000 datos del clima.

```python
cl_AUS=climaAUS_df[0:500]
cl_AUS
```

Grafica con pyspc de la columna de temperatura máxima de los primeros 500 días desde el 2009:

```python
a = spc(cl_AUS.MaxTemp) + ewma()
print(a)
```

Grafica con pyspc de la columna de temperatura mínima de los primeros 500 días desde el 2009:

```python
b = spc(cl_AUS.MinTemp) + ewma()
print(b)
```
